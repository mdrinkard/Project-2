```python
dict_info ={
    'Variable Name':'Description',
    'LoanNr_ChkDgt':	'Identifier Primary key',
    'Name': 'Borrower name',
    'City':'Borrower city',
    'State': 'Borrower state',
    'Zip': 'Borrower zip code',
'Bank' : 'Bank name',
'BankState':	'Bank state',
'NAICS' :'North American industry classification system code',
'ApprovalDate':'Date SBA commitment issued',
'ApprovalFY': 'Fiscal year of commitment',
'Term': 'Loan term in months',
'NoEmp' :	'Number of business employees',
'NewExist': '1 = Existing business, 2 = New business',
'CreateJob':	'Number of jobs created',
'RetainedJob' :	'Number of jobs retained',
'FranchiseCode': 'Franchise code, (00000 or 00001) = No franchise',
'UrbanRural' :	'1 = Urban, 2 = rural, 0 = undefined',
'RevLineCr':	'Revolving line of credit: Y = Yes, N = No',
'LowDoc':	'LowDoc Loan Program: Y = Yes, N = No',
'ChgOffDate': 'The date when a loan is declared to be in default',
'DisbursementDate' : 	'Disbursement date',
'DisbursementGross' :	'Amount disbursed',
'BalanceGross' : 'Gross amount outstanding',
'MIS_Status' : 'Loan status charged off = CHGOFF, Paid in full =PIF',
'ChgOffPrinGr':	'Charged-off amount',
'GrAppv':	'Gross amount of loan approved by bank',
'SBA_Appv': 'SBA’s guaranteed amount of approved loan'
}
```

```python
### Defining a Function to report the qualities of each of the columns of the dataframe
def report(df):
    col = []
    d_type = []
    uniques = []
    n_uniques = []
    nan = []

    for i in df.columns:
        col.append(i)
        d_type.append(df[i].dtypes)
        uniques.append(df[i].unique()[:5])
        n_uniques.append(df[i].nunique())
        nan.append(df[i].isna().sum()/len(df)*100)

    return pd.DataFrame({'Column': col, 'd_type': d_type, 'unique_sample': uniques, 'n_uniques': n_uniques, 'nan%': nan})

report(df)
```

```python
### Deciding what columns to drop
### We are trying to predict whether or not to give a business a Loan
### Relevant colunmns are MIS_Status, Lowdoc, RevLinecr, UrbanRural, Franchise Code, Retained Job, Create_job, NewExist, NoEmp, Term, NAICS, BankState, City, State. all others we will drop.
dict_info
```

```python
### These Columns do not add much to the model or will leak data to the neural network algorithm and so I am dropping.
columns_to_drop = ['LoanNr_ChkDgt', 'Name', 'Zip','ApprovalDate','ApprovalFY','ChgOffDate','DisbursementDate','DisbursementGross','GrAppv','SBA_Appv','ChgOffPrinGr','BalanceGross']
df_adjusted = df.copy()
df_adjusted
```

```python
df_adjusted = df_adjusted.drop(columns=columns_to_drop)
df_adjusted
```

```python
### We have dropped the irrelavant columns and now we will adjust the following columns to allow the Neural Network to function.
### First Column to work on will be the MIS_Status, this column shows whether or not the business defaulted on the Loan and will function as the Y column for the model.
df_adjusted = df_adjusted.dropna(subset='MIS_Status')
df_adjusted['MIS_Status'].value_counts(dropna=False)
```

```python
### Currently 2 Variables in the MIS_Status columns P I F and CHGOFF meaning paid or default. Changing it to an integer 0 and 1.
df_adjusted['MIS_Status'] = np.where(df_adjusted['MIS_Status']== 'P I F',0,1)
```

```python
df_adjusted['MIS_Status'].value_counts()
```

```python
### Working on the NAICS Column now. This indicates industry, however, i'd like to assign each id to the actual sector of industry it is located in.
### Establishing a dictionary to act as a guide
naics_to_sector = {
    '0' : 'Unknown Sector',
    '11': 'Agriculture, Forestry, Fishing and Hunting',
    '21': 'Mining, Quarrying, and Oil and Gas Extraction',
    '22': 'Utilities',
    '23': 'Construction',
    '31': 'Manufacturing',
    '32': 'Manufacturing',
    '33': 'Manufacturing',
    '42': 'Wholesale Trade',
    '44': 'Retail Trade',
    '45': 'Retail Trade',
    '48': 'Transportation and Warehousing',
    '49': 'Transportation and Warehousing',
    '51': 'Information',
    '52': 'Finance and Insurance',
    '53': 'Real Estate and Rental and Leasing',
    '54': 'Professional, Scientific, and Technical Services',
    '55': 'Management of Companies and Enterprises',
    '56': 'Administrative and Support and Waste Management and Remediation Services',
    '61': 'Educational Services',
    '62': 'Health Care and Social Assistance',
    '71': 'Arts, Entertainment, and Recreation',
    '72': 'Accommodation and Food Services',
    '81': 'Other Services (except Public Administration)',
    '92': 'Public Administration',
    # Add more NAICS codes and sectors as needed
}
```

```python
### in the NAICS column, only the first 2 digits are required to indicate the industry it is in.
df_adjusted['NAICS'] = df_adjusted['NAICS'].astype(str).str[:2]
```

```python
###adding a column to map industry to the number indicated above
df_adjusted['Industry'] = df_adjusted['NAICS'].map(naics_to_sector)
```

```python
### Dropping original NAICS column as the data is I want is reflected in the industry column.
df_adjusted = df_adjusted.drop(columns='NAICS')
```

```python
### Starting with the Term Column indicating Loan Term I'll be breaking it up by dividing it into quartiles of the year.
### I'll start by adding a column to adjust the data where I want it adjusted.
df_adjusted['TermGroup'] = np.where(df_adjusted['Term']<=90,'Below 3 months',
                            np.where((df_adjusted['Term']>90) & (df_adjusted['Term']<=180), '3-6 months',
                            np.where((df_adjusted['Term']>180) & (df_adjusted['Term']<=365), '6-12 months', 'More Than a Year')))
display(df_adjusted[['Term', 'TermGroup']].sample(10))
```

```python
### Dropping Term Column as the Data I want is reflected in the column TermGroup
df_adjusted = df_adjusted.drop(columns='Term')
```

```python
### Filtering out Null values from LowDoc Columns
df_adjusted = df_adjusted[df_adjusted['LowDoc'].isin(['N','Y'])]
```

```python
### RevLineCr has alot of values not Y and N. I believe the 0 and T values indicate no and yes respectively. so I will be making the assumption and changing those values to N and Y.
df_adjusted['RevLineCr'].value_counts(dropna=False)
```

```python
df_adjusted['RevLineCr'] = df_adjusted['RevLineCr'].replace({'0': 'N', 'T': 'Y'})
```

```python
df_adjusted = df_adjusted[df_adjusted['RevLineCr'].isin(['N','Y'])]
display(df_adjusted['RevLineCr'].value_counts(dropna=False))
```

```python
### Dropping nulls from the columns I plan to categorize
df_adjusted.dropna(subset=['Bank' , 'BankState','City'], inplace=True)
```

```python
df_adjusted.shape
```

```python
df_adjusted_2 = df_adjusted.copy()
```

```python
count_bankst = df_adjusted['BankState'].value_counts().to_frame().reset_index()
count_bankst['index'] = count_bankst['index'].astype(str)
```

```python
count_bankst = count_bankst.rename(columns={'BankState': 'BankstCount'})
count_bankst.dtypes
```

```python
df_adjusted_2 = df_adjusted_2.merge(count_bankst, left_on='BankState', right_on='index', how='left')
```

```python
df_adjusted_2['NewBankState'] = np.where(df_adjusted_2['BankstCount'] > 15, df_adjusted_2['BankState'], 'OTHER')
```

```python
df_adjusted_2.drop(['BankState', 'BankstCount', 'index'], axis=1, inplace=True)
```

```python
###Dropping nulls from Franchise Column
df_adjusted_2['IsFranchise'] = np.where(((df_adjusted_2['FranchiseCode']==1) | (df_adjusted_2['FranchiseCode']==0)), 'Y', 'N')

```

```python
df_adjusted_2.drop(columns='FranchiseCode', inplace=True)
```

```python
### Working on City Column
count_city = df_adjusted_2['City'].value_counts().to_frame().reset_index()
count_city.rename(columns={'City':'City_Count','index' : 'City'}, inplace=True)
```

```python
count_city[(count_city['City_Count']>25)&(count_city['City_Count']<=50)]
```

```python
df_adjusted_2 = df_adjusted_2.merge(count_city, on='City')
```

```python
df_adjusted_2.shape
```

```python
df_adjusted_2['NewCity'] = np.where(df_adjusted_2['City_Count']>100, df_adjusted_2['City'], 'OTHER')
df_adjusted_2.drop(['City','City_Count'],axis=1, inplace=True)
```

```python
df_adjusted_3= df_adjusted_2.copy()
```

```python
### Working on Banks now
count_bank = df_adjusted_3['Bank'].value_counts().to_frame().reset_index()
count_bank = count_bank.rename(columns={'Bank': 'Bank_Count', 'index' : 'Bank'})
```

```python
df_adjusted_3 = df_adjusted_3.merge(count_bank, on='Bank')

```

```python
df_adjusted_3['NewBank'] = np.where(df_adjusted_3['Bank_Count']>5000, df_adjusted_3['Bank'], 'OTHER')
```

```python
df_adjusted_3['Bank'].value_counts(dropna=False)
```

```python
df_adjusted_3.drop(columns=['Bank','Bank_Count'], axis=1, inplace=True)
```
